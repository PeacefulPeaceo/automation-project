## 1. Calculate Total Sales for the Day: from sales data for morning and evening shifts, produce total sales for the day.

## 2. Calculate Worker's Salary: given hourly rate and hours worked by a worker. retrieve the worker's salary.

```python
def workers_salary():
    '''
    Takes in an input of the number of hours a worker worked and calculates the total salary using a fixed hourly rate
    and returns the total salary.
    '''
    try:
#         enter the number of hours worked
        rate_input = input('enter the salary per hour: ')
    
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
        while rate_input.isnumeric() == False:
            print('wrong input. takes only numeric values')
            rate_input = input('enter the salary per hour: ')
        rate_per_hour = float(rate_input)
            
        hours_input = input('enter the number of hours you worked: ')
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
        while hours_input.isnumeric() == False:
            print('wrong input. takes only numeric values')
            hours_input = input('enter the number of hours you worked: ')
        hours_worked = float(hours_input)
            
        worker_salary = rate_per_hour * hours_worked
        print(f' Your salary for today is: ${worker_salary}')
        return
    except:
        print('invalid entry. Kindly enter a numeric value.')
```

## 3. Calculate Profit: given a list of numbers representing total sales and total cost of items sold, find the profit.(or loss if negative)

```python
def profit_or_loss_made():
    
    '''takes an input of the selling price and cost price of all the items sold, converts it into a list, 
        sum it up and calculates the profit or loss'''
    
    while True:
    
        try:
#             enter the costost and selling prices
            sales = input('enter the total sales prices of each item sold separating them with a comma: ')
            cost = input('enter the total cost price of each item in your sales list separating them with a comma: ')
            
#             use .split() method to return a list of the sales and cost input
            lst_sales = sales.split(',')
            lst_cost = cost.split(',')

#             using list comprehension. write a function that convert the string values to float
            lst_sellingprice = [float(num) for num in lst_sales]
            lst_costprice = [float(num) for num in lst_cost]

            print('\n')
            print('Your sales:' ,lst_sellingprice)
            print('your cost: ',lst_costprice)

#             get the sum of the sales and cost lists
            total_sp = sum(lst_sellingprice)
            total_cp = sum(lst_costprice)

            print('\n')
            print(f'Your total sales is: ${total_sp}')
            print(f'Your total costprice of goods sold is: ${total_cp}')
            print('Wait while we calculate your profit or loss......')

#             use the comparison operator to determine if you are to calculate profit or loss based on two conditions
            if total_sp > total_cp:
                profit = total_sp - total_cp
                print(f'You made a profit of ${profit}')
                
            else:
                loss = total_cp - total_sp
                print(f'You made a loss of ${loss}')
            return
        except:
            print('invalid input. your input only takes numbers.')
    else:
        sales = input('enter your individual sales prices separating them with a comma: ')
        cost = input('enter the cost of the goods in your sales separating them with a comma: ')
```

## 4. Calculate Tips for a Shift: from sales data for a specific shift, workers get tipped for the shift (2% of shift sales).

```python
def shift_tips():
    '''Takes an input of the shift and total sales for that shift and calculates the tip for the shift '''
    
    while True:
        try:
    #         takes an input of the shift and stores it as shift
            shift_option = int(input('Kindly choose your shift, Enter (1) for Morninig/ enter (2) for Evening: '))
            shift = 'Morning'
            
#      use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
            while shift_option not in (1,2):
                print('wrong input. enter 1 or 2')
                shift_option = int(input('Kindly choose your shift, Enter (1) for Morninig/ enter (2) for Evening: '))
                
                
            if shift_option == 2:
                shift = 'Evening'

    #         use a while loop to check for correct input. the user should input either morning or evening

            if shift[0].upper() == 'M':


                morn_sales = input('enter the total sales prices of each item sold separating them with a comma: ')
                


    #             use .split() method to return a list of the sales and cost input
                lst_sales = morn_sales.split(',')

    #             using list comprehension, write a function that convert the string values to float and sums the list
                lst_morn_sales = [float(num) for num in lst_sales]
                morning_sales = sum(lst_morn_sales)

                tip = 2/100 * morning_sales
                print('\n')
                print(f'your tip for morning shift is ${tip}')
                return

            elif shift[0].upper() == 'E':
                eve_sales = input('enter the total sales prices of each item sold separating them with a comma: ')


    #             use .split() method to return a list of the sales and cost input
                lst_sales2 = eve_sales.split(',')

    #             using list comprehension, write a function that convert the string values to float and sums the list
                lst_eve_sales = [float(num) for num in lst_sales2]
                evening_sales = sum(lst_eve_sales)

    #                 calculate the tip by multiplying the evening sales by 2%
                tip = round((2/100 * evening_sales),2)
                print('\n')
                print(f'your tip for evening shift is ${tip}')
                return

        except:
            print('invalid entry, please try again by entering a numerical value for your total sales.')
            
    else:
        shift_option = int(input('Kindly choose your shift, Enter (1) for Morninig/ enter (2) for Evening: '))
        
```

## 5. Calculate Total Tips for the Day: with sales data for morning and evening shifts, return total tips for the day (sum of tips from both shifts).

```python
def total_tips():
    ''' takes an input of the morning and evening sales, stores it in didfferent lists, 
        calculates their tips individually and prints the sum of the tips '''
    
    try:
#               create an empty list that will hold the sales values for morning shift
        lst_morn_sales = []

#             input the number of sales you want to enter
        num_input = input('how many sales do you want to enter for Morning shift? ')
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
        while num_input.isnumeric() == False:
            num_input = input('how many sales do you want to enter for Morning shift? ')
        num = int(num_input)

        for i in range(num):
            price = input('Enter your sales value: ')
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
            while price.isnumeric() == False:
                print('wrong input. takes only numeric values')
                price = input('Enter your sales value: ')
            lst_morn_sales.append(float(price))


#              create an empty list that will hold the sales values for evening shift
        lst_eve_sales = []

#             input the number of sales you want to enter
        num_input2 = input('how many sales do you want to enter for Evening shift? ')
    
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
        while num_input2.isnumeric() == False:
            num_input2 = input('how many sales do you want to enter for Evening shift? ')
        num2 = int(num_input2)
        
        for v in range(num2):
            price2 = input('Enter your sales value: ')
            
#     use a while loop to check if the input is a numeric string. If it is false, keep asking to enter asking numeric value
            while price2.isnumeric() == False:
                print('wrong input. takes only numeric values')
                price2 = input('Enter your sales value: ')
            lst_eve_sales.append(float(price2))

        print('\n')
        print('morning sales:' ,lst_morn_sales)
        print('evening sales: ',lst_eve_sales)

#             get the sum of the sales and cost lists
        total_morn_sales = sum(lst_morn_sales)
        total_eve_sales = sum(lst_eve_sales)
        
#         calculates the tips separately
        morning_tip =  round((2/100 * total_morn_sales),2)
        evening_tip = round((2/100 * total_eve_sales),2)
        
#         adds the result of the tips and displays it
        total_tip = morning_tip + evening_tip
        print('\n')
        print(f'the total tip for today is ${total_tip}')
        return
    
    except:
        print('invalid entry. Kindly enter a numerical value.')
        
```

## User Interface:

Create a user-friendly interface that displays a menu of available operations.
Accept user input to choose the desired operation (1-5) or exit (6).


```python
def user_interface():
    ''' while true, keep runing. Display a menu for user to select from 
            and the function attached to the number chosen number will be called '''
    
    while True:
        print('\n')
        print('Business Operations\n')
        print('1. Calculate total sales')
        print('2. Calculate worker salary')
        print('3. Calculate profit or loss')
        print('4. Calculate tips for your shift')
        print('5. Calculate total shift for the day')
        print('6. exit')
        print('\n')
        
        choice = input('Select your desired option (1-6): ')
        
        if choice == '1':
            print('\n')
            print('-------------------- Calculate total sales ---------------------\n')
            total_sales()
        elif choice == '2':
            print('\n')
            print('-------------------- Calculate worker salary ---------------------\n')
            workers_salary()
        elif choice == '3':
            print('\n')
            print('-------------------- Calculate profit or loss ---------------------\n')
            profit_or_loss_made()
        elif choice == '4':
            print('\n')
            print('-------------------- Calculate tips for your shift ---------------------\n')
            shift_tips()
        elif choice == '5':
            print('\n')
            print('-------------------- Calculate total shift for the day ---------------------\n')
            total_tips()
        elif choice == '6':
            print('\n')
            print('exiting.............')
            print('Thank you for using this service')
            break
        else:
            print('invalid entry, choose a number between 1 and 6')     
```
